```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# d = 2

## равномерное распределение

```python
size = 100
x = np.random.uniform(0,10,(size,2))
x = x[x[:,0].argsort()]
# rule = x[:,0] * np.sin(0.4*x[:,0]) + np.exp(0.25 * x[:,0] )
rule =  np.sin(2 * x[:,0]) - np.exp(0.25 * x[:,0] ) + 10
# rule = rule[np.argsort(rule,1)]
# plt.scatter(x[:,0],x[:,1])
# plt.plot(x[:,0], rule, color='red')
# plt.ylim(0,10)
# plt.show()

data = pd.DataFrame()
data['X'] = x[:,0]
data['Y'] = x[:,1]
data['true_rule'] = rule
data['Class'] = 0
data.loc[(data['Y'] > data['true_rule']) ,'Class'] = 1

plt.scatter(data.loc[data['Class'] == 0, 'X'],data.loc[data['Class'] == 0, 'Y'], color='green')
plt.scatter(data.loc[data['Class'] == 1, 'X'],data.loc[data['Class'] == 1, 'Y'], color='magenta')
plt.ylim(0,10)
plt.plot(data['X'], data['true_rule'], color='red')
plt.show()
```

```python
data.to_csv('RandomUniformD2.csv',index=False)
```

# Гауссовы сгустки

```python
size = 100
# x_0 = np.random.normal(3,0.7,size)
# y_0 = np.random.normal(3,0.7,size)

# x_1 = np.random.normal(5,1.5,size)
# y_1 = np.random.normal(7,0.5,size)

x_0 = np.random.normal(3,0.7,size)
y_0 = np.random.normal(3,0.7,size)

x_1 = np.random.normal(8,0.5,size)
y_1 = np.random.normal(6,2,size)

plt.scatter( x_0, y_0, color='magenta' )
plt.scatter( x_1, y_1, color='green' )
plt.ylim(0,10)
plt.xlim(0,10)
plt.show()

data0 = pd.DataFrame()
data0['X'] = x_0
data0['Y'] = y_0
data0['Class'] = 0

data1 = pd.DataFrame()
data1['X'] = x_1
data1['Y'] = y_1
data1['Class'] = 1

data = pd.concat( [data0, data1], ignore_index=True )
data = data.sample(frac = 1)

plt.scatter(data.loc[data['Class'] == 0, 'X'],data.loc[data['Class'] == 0, 'Y'], color='magenta')
plt.scatter(data.loc[data['Class'] == 1, 'X'],data.loc[data['Class'] == 1, 'Y'], color='green')
plt.ylim(0,10)
plt.xlim(0,10)
plt.show()
```

```python
data.to_csv('TwoGaussD2.csv',index=False)
```

## d = 5

## Равномерное распределение

```python
size = 100
x = np.random.uniform(0,10,(size,5))

rule = (x[:,0] + 0.25 * x[:,1] - x[:,2]) / 0.25 * x[:,3]

data = pd.DataFrame()
data['X0'] = x[:,0]
data['X1'] = x[:,1]
data['X2'] = x[:,2]
data['X3'] = x[:,3]
data['Y'] = x[:,4]
data['true_rule'] = rule
data['Class'] = 0
data.loc[(data['Y'] > data['true_rule']) ,'Class'] = 1

for column in data.columns[:-3]:
    plt.scatter(data.loc[data['Class'] == 0, column],data.loc[data['Class'] == 0, 'Y'], color='green')
    plt.scatter(data.loc[data['Class'] == 1, column],data.loc[data['Class'] == 1, 'Y'], color='magenta')
    plt.ylim(0,10)
    plt.scatter(data[column], data['true_rule'], color='red')
    plt.xlabel(column)
    plt.ylabel("Y")
    plt.show()
```

```python
data.to_csv('RandomUniformD5.csv',index=False)
```
