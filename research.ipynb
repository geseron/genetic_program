```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats 
```

# d = 2

## Равномерное

```python
data_prop =  pd.read_csv("exp/D2/UniformD2/proportional.csv", index_col=0)
data_tour = pd.read_csv("exp/D2/UniformD2/tournament.csv", index_col=0)
print(f"Пропорциональная")
print(data_prop.describe())
print(f"Турнирная")
print(data_tour.describe())
```

```python
sns.histplot(data_prop['f1'], kde=True, stat="density", color='orange', bins=20, label ='Пропорциональная')
sns.histplot(data_tour['f1'], kde=True, stat="density", color='black', bins=20, label ='Турнирная')  
plt.title('Плотности вероятностей (равномерное d=2)')  
plt.xlabel('f1')  
plt.ylabel('Плотность вероятности')  
plt.legend()

plt.tight_layout()  
plt.show()  
```

$H_0:$ выборки равны

```python
u_statistic, p_value = stats.mannwhitneyu(data_prop['f1'], data_tour['f1'])  

print(f"U-statistic: {u_statistic}, p-value: {p_value}")  

# Уровень значимости  
alpha = 0.05  
if p_value < alpha:  
    print("Отвергаем нулевую гипотезу. Выборки статистически различаются.")  
else:  
    print("Нет оснований отвергать нулевую гипотезу.")  
```
